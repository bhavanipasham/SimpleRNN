```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##sentences
sentences = ['the cat sat on the mat',
             'the dog sat on the log',
             'the cat sat on the log',
             'The dog is barking']
sentences
```

```python
##Define vocabulary size
vocab_size = 1000
```

```python
###one hot representation of sentences
one_hot_representations = [one_hot(word, vocab_size) for word in sentences]
one_hot_representations
```

```python
sentences
```

```python
##word embedding reprensentation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 5
embedded_docs = pad_sequences(one_hot_representations, padding='pre', maxlen=sent_length)
```

```python
print(embedded_docs)
```

```python
##feature representation using embedding layer
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
